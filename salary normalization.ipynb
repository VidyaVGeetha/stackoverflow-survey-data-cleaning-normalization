# ----------------------Section 3: Normalizing Compensation Data-------------------

```python
# Step 1: Identify Compensation-Related Columns
comp_cols = [col for col in df.columns if "Comp" in col or "Salary" in col]
print(comp_cols)
```

```python
# Step 2: Fill missing values with a suitable statistic
median_salary = df["ConvertedCompYearly"].median()
df["ConvertedCompYearly"].fillna(median_salary, inplace=True)
```

Before normalizing the ConvertedCompYearly column, missing values were handled to prevent errors. The missing values were either removed or filled using a suitable value such as the median. After cleaning, the column was then used for compensation normalization.

Used median (recommended for salary, because data contains outliers):

```python
# Step 3: Normalize ConvertedCompYearly using Min-Max Scaling.

# 3.1:  Find the smallest and largest salary 
min_val = df["ConvertedCompYearly"].min() # minimum salary
max_val = df["ConvertedCompYearly"].max() # maximum salary

print("Minimum Salary:", min_val)
print("Maximum Salary:", max_val)
```

```python
# 3.2: Apply the Min-Max formula
df["ConvertedCompYearly_MinMax"] = (
    df["ConvertedCompYearly"] - min_val
) / (max_val - min_val)
```

* Big salaries get values closer to 1
*  Small salaries get values closer to 0
*  Everything else falls in between
* This makes comparison fair and scaled.

```python
# 3.3: Check summary statistics
df["ConvertedCompYearly_MinMax"].describe()
```

## Key Insihgts
* The ConvertedCompYearly column was successfully normalized using Min-Max scaling.
* ConvertedCompYearly_MinMax is the new column name.
* The calculated values are saved into the new column.
* The values now range from 0 to 1, which confirms normalization worked correctly.
* Most values are closer to 0, meaning many salaries are on the lower end of the scale.
* This happens because a few very high salaries (outliers) push the scale upward.

```python
# 3.4: Apply Z-score Normalization to ConvertedCompYear
mean_val = df["ConvertedCompYearly"].mean()
std_val = df["ConvertedCompYearly"].std()

df["ConvertedCompYearly_Zscore"] = (df["ConvertedCompYearly"] - mean_val) / std_val

# 3.5: Verify
df["ConvertedCompYearly_Zscore"].describe()
```

## Key Insihgts
* The mean is effectively 0. (The mean value appears as a very small negative number due to floating-point rounding)
* standard deviation is 1
* Z-score normalization worked correctly


# ------------Section 4: Visualization of Normalized Data------------

```python
# Step 1: Visualize Salary Before & After Normalization
# Plot All Three Distributions
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

# Original Salary Distribution
plt.subplot(1,3,1)
plt.hist(df["ConvertedCompYearly"], bins=30)
plt.title("Original Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

# Min-Max Normalized Salary
plt.subplot(1,3,2)
plt.hist(df["ConvertedCompYearly_MinMax"], bins=30)
plt.title("Min-Max Normalized Salary")
plt.xlabel("0 - 1 Scale")

# Z-Score Normalized Salary
plt.subplot(1,3,3)
plt.hist(df["ConvertedCompYearly_Zscore"], bins=30)
plt.title("Z-Score Normalized Salary")
plt.xlabel("Standardized Scale")

plt.tight_layout()
plt.show()
```

```python


#  Create a copy without extreme outliers (for clearer charts)
df_small = df[df["ConvertedCompYearly"] < df["ConvertedCompYearly"].quantile(0.95)]

plt.figure(figsize=(14,5))

# -----------------------
# 1.1: Original Salary

plt.subplot(1,3,1)
plt.hist(df_small["ConvertedCompYearly"], bins=30, color="#4C72B0", edgecolor="black")
plt.title("Original Salary Distribution", fontsize=12, fontweight="bold")
plt.xlabel("Salary", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.4)

# -------------------------------
# 1.2: Min-Max Normalized Salary

plt.subplot(1,3,2)
plt.hist(df_small["ConvertedCompYearly_MinMax"], bins=30, color="#55A868", edgecolor="black")
plt.title("Min-Max Normalized Salary", fontsize=12, fontweight="bold")
plt.xlabel("0–1 Scale", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.4)


# 1.3: Z-Score Normalized Salary

plt.subplot(1,3,3)
plt.hist(df_small["ConvertedCompYearly_Zscore"], bins=30, color="#C44E52", edgecolor="black")
plt.title("Z-Score Normalized Salary", fontsize=12, fontweight="bold")
plt.xlabel("Standardized Scale", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```
